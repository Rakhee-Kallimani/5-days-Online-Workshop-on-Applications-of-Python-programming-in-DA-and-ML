3. Write a Pandas program to add, subtract, multiple and divide two Pandas Series and print their results using the Sample Series: [2, 4, 6, 8, ...100], [1, 3, 5, 7, 9...99]

```python
import pandas as pd
ds1 = pd.Series([i for i in range(2,101,2)])
ds2 = pd.Series([i for i in range(1,100,2)])
ds = ds1 + ds2
print("Add two Series:")
print(ds)
print("\nSubtract two Series:")
ds = ds1 - ds2
print(ds)
print("\nMultiply two Series:")
ds = ds1 * ds2
print(ds)
print("\nDivide Series1 by Series2:")
ds = ds1 / ds2
print(ds)
```

4. Write a Pandas program to compare (Equivalence, greater than, less than, not equal to )the elements of the two Pandas Series.
#Sample Series: [All positive numbers divisable by 7 upto 1000], [All positive numbers divisable by 11 upto 1000]


```python
import pandas as pd
a = [i for i in range(1000) if i%7 == 0]
b = [i for i in range(1000) if i%11 == 0]
ds1 = pd.Series(a)
ds2 = pd.Series(b)
print("Series1:")
print(ds1)
print("\nSeries2:")
print(ds2)

if len(a) != len(b):
    if len(a)> len(b):
        a= a[:len(b)]
    else:
        b= b[:len(a)]
        
ds1 = pd.Series(a)
ds2 = pd.Series(b)
print("\nUpdated Series1:")
print(ds1)
print("\nUpdated Series2:")
print(ds2)
print("\nCompare the elements of the said Series:")
print("\nEquals:")
print(ds1 == ds2)
print("\nGreater than:")
print(ds1 > ds2)
print("\nLess than:")
print(ds1 < ds2)
print("\nNot Equals:")
print(ds1 != ds2)
```

8. Write a Pandas program to create a subsets of a given series based on value and condition.
#Sample NumPy array: d1 = [10, 20, 30, 40, 50... 1000]
#value and condtion :  Subset 1 :numbers divisable by 7  Subset 2 : numbers divisable by 5

```python
import pandas as pd
s = pd.Series([i for i in range(10,1001,10)])
print("Original Data Series:")
print(s)
print("\nSubset 1 :numbers divisable by 7 :")
new_s1 = s[s%7==0]
print(new_s1)
print("\nSubset 2 :numbers divisable by 5 :")
new_s2 = s[s%5==0]
print(new_s2)
```

9. Write a Pandas program to create the mean and standard deviation, maximum and minimum of the data of a given Series of the prime numbers upto 1000

```python
import pandas as pd
import sympy
s = pd.Series(sympy.primerange(0, 1000))
print("Original Data Series:")
print(s)
print("\nMean of the Series:")
print(s.mean())
print("\nStandard deviation of the Series:")
print(s.std())
print("\nMaximum of the Series:")
print(s.max())
print("\nMinimum of the Series:")
print(s.min())
```
